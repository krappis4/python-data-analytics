```python
import numpy as np
import regex as re
import matplotlib.pyplot as plt
```

```python
norm_ = np.random.randint(0, 1000, size=10000)
plt.scatter([i for i in range(10000)], norm_, alpha=0.3)
```

```python
def win_(n):
  if re.fullmatch(r'^\d{2}0$', str(n)):
    return 1
  if re.fullmatch(r'^\d00$', str(n)):
    return 2
  if re.fullmatch(r'^\d{2}7$', str(n)):
    return 3
  if re.fullmatch(r'^\d77$', str(n)):
    return 5
  if n == 111:
    return 10
  if n == 333:
    return 15
  if n == 555:
    return 50
  if n == 999:
    return 100
  if n == 777:
    return 200
  return 0
def test_():
  dt = []
  input_ = np.random.randint(0, 1000, size=10000)
  summ = 0
  for i in input_:
    summ += (win_(i) - 1)
    dt.append(win_(i))
  plt.plot([i for i in range(10000)], dt)
  print(summ, summ/10000)
  return summ/10000
ans = 0
for j in range(5):
  for i in range(2):
    ans += test_()
  plt.show()
print('Среднее значение прибыли за игру для 100000', ans / 10)
```

Я прогнала тестовые данные на 1000000 игр, на выходе вышло, что в среднем за игру игрок теряет 47 копеек, что очевидно для него не выгодно. (-0.4694720000000001)
