```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import seaborn as sns
from scipy.stats import pearsonr
import cufflinks
cufflinks.go_offline()
%matplotlib inline
import statistics
import ssl
from decimal import Decimal
from sklearn import preprocessing
from plotly.offline import iplot
```

```python
dt = pd.read_csv('2015.csv')
dt.plot(y='Economy (GDP per Capita)', x='Happiness Score', kind='line')
```

```python
d = dt[['Country', 'Health (Life Expectancy)']].groupby('Country').sum().sort_values(by='Health (Life Expectancy)', ascending=False).head(10)
d
plt.pie(d, labels = dt['Country'].head(10), shadow = True, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
np_scaled1 = min_max_scaler.fit_transform(dt[['Freedom', 'Happiness Score', 'Trust (Government Corruption)']])
df = pd.DataFrame(np_scaled1)
df.plot.box(figsize=(5, 3))
```

```python
sns.heatmap(dt[['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']].corr(),annot=True,cmap='RdYlGn',linewidths=0.2) #data.corr()--> 
# gcf() -   
fig=plt.gcf()
# set_size_inches() -   
fig.set_size_inches(12,12)
plt.show()
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

np_scaled2 = min_max_scaler.fit_transform(dt[['Family', 'Health (Life Expectancy)', 'Happiness Score']])
#np_scaled2 = min_max_scaler.fit_transform(dt[['Freedom', 'Trust (Government Corruption)']])
dt_norm = pd.DataFrame(np_scaled2)
y1 = dt_norm[0]
y2 = dt_norm[1]
y3 = dt_norm[2]
x1 = [i for i in range(len(y1))]
x2 = [i for i in range(len(y2))]
x3 = [i for i in range(len(y3))]
plt.scatter(x1, y1,color = 'red',label = 'Family', alpha=0.3)
#   ,  ,  
plt.scatter(x2, y2, color = 'blue', label = 'Health', alpha=0.3)
plt.scatter(x3, y3, color = 'green', label = 'Happiness Score', alpha=0.3)
plt.legend(loc = 'best')
```
