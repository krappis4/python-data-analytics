```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import seaborn as sns
from scipy.stats import pearsonr
import cufflinks
# hello, vita!
cufflinks.go_offline()
%matplotlib inline
import statistics
import ssl
from decimal import Decimal
from sklearn import preprocessing
from plotly.offline import iplot
```

```python
def f(x):
    return np.sin(x)  # sin(x)
```

```python
def integral_sum(f, a, b, n, method='right'):
    dx = (b - a) / n
    x = np.linspace(a, b, n+1)
    if method == 'left':
        x_osn = x[:-1]
    elif method == 'right':
        x_osn = x[1:]
    elif method == 'middle':
        x_osn = (x[:-1] + x[1:]) / 2
    elif method == 'random':
        x_osn = np.random.uniform(low=a, high=b, size=n)
    else:
        raise ValueError("Invalid method! Choose from 'left', 'right', 'middle', or 'random'.")

    integral = np.sum(f(x_osn) * dx)
    y_osn = [f(x_osn[i]) for i in range(len(x_osn))]
    return integral, x, f(x), x_osn, y_osn
```

```python
def print_(mode):
  a = 0
  b = 2 * np.pi
  n1 = 10
  n2 = 100
  method = mode
  integral, x, y, x_osn, y_osn = integral_sum(f, a, b, n1, method)
  integral2, x2, y2, x_osn2, y_osn2 = integral_sum(f, a, b, n2, method)

  print(f"Значение интеграла: {integral}")
  print(f"Значение интеграла 2: {integral2}")


  plt.plot(x2, y2, label='sin(x)', color = "red")
  plt.bar(x_osn2, y_osn2, color="pink", edgecolor="red", alpha = 0.25, width=0.06)
  #plt.fill_between(x2[:-1], 0, y2[:-1], alpha=0.3)
  plt.plot(x, y, label='sin(x)', color="black")
  plt.bar(x_osn, y_osn, color="blue", edgecolor="darkblue", alpha = 0.25, width=0.63)
  for i in x_osn:
    plt.axvline(i, color='purple', linestyle='--')
  #plt.fill_between(x[:-1], 0, y[:-1], alpha=0.3)
  plt.title('График функции и интегральной суммы')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.legend()
  plt.grid(True)
  plt.show()
```
